```python
!pip install ipywidgets
import ipaddress
import ipywidgets as widgets
from IPython.display import display, clear_output
def calculate_ip_details(ip_input):
    try:
        network = ipaddress.ip_network(ip_input, strict=False)

        first_octet = int(str(network.network_address).split('.')[0])
        if 1 <= first_octet <= 126:
            network_class = "A"
        elif 128 <= first_octet <= 191:
            network_class = "B"
        elif 192 <= first_octet <= 223:
            network_class = "C"
        elif 224 <= first_octet <= 239:
            network_class = "D (Multicast)"
        elif 240 <= first_octet <= 255:
            network_class = "E (Experimental)"
        else:
            network_class = "Unknown"

        result = {
            "Network Address": str(network.network_address),
            "Broadcast Address": str(network.broadcast_address),
            "Subnet Mask": str(network.netmask),
            "Number of Usable Hosts": network.num_addresses - 2 if network.num_addresses > 2 else 0,
            "First Usable IP": str(list(network.hosts())[0]) if network.num_addresses > 2 else "N/A",
            "Last Usable IP": str(list(network.hosts())[-1]) if network.num_addresses > 2 else "N/A",
            "Network Class": network_class
        }
        return result
    except Exception as e:
        return {"Error": str(e)}
ip_input = widgets.Text(
    value='192.168.1.0/24',
    placeholder='Enter IP/CIDR (e.g., 192.168.1.0/24)',
    description='IP/CIDR:',
    disabled=False
)

button = widgets.Button(
    description='Calculate',
    button_style='success',
    tooltip='Click to calculate IP details',
    icon='check'
)
output = widgets.Output()
def on_button_clicked(b):
    with output:
        clear_output()
        result = calculate_ip_details(ip_input.value)
        for key, value in result.items():
            print(f"{key}: {value}")
button.on_click(on_button_clicked)
display(ip_input, button, output)
```
